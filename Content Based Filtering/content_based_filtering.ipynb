## Content Based Recommendation System

Now lets make a recommendations based on the movie's plot summaries given in the averview column. So if our user gives us a movie title, our goal is to recommend movies that share similar plot summaries

```python
#now we calculate TF-IDF value for each movie
# import required module
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,max_features=None, strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',ngram_range=(1,3),stop_words='english')

#filling Nans with empty string
movie_df_merge['overview']=movie_df_merge['overview'].fillna('')
```

```python
#fitting the TF-IDF on the 'overview' text
tfv_matrix =tfv.fit_transform(movie_df_merge['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
#compute the sigmoid kernel
sig = sigmoid_kernel (tfv_matrix,tfv_matrix)
```

```python
sig[0]
```

```python
#reverse mapping of ind
indices = pd.Series(movie_df_merge.index,index=movie_df_merge['original_title']).drop_duplicates()
```

```python
indices
```

```python
def get_recommendation(title,sig=sig):
    #getting the index of moive
    if "Spy Kid" not in indices:
        return "Movie Name not found"
    indx = indices[title]

    #get the pairwise similarity score for indx
    sig_scores=list(enumerate(sig[indx]))

    #sort the movies based on similarity score
    sig_scores =sorted(sig_scores,key=lambda x: x[1], reverse=True)

    #get the index of top 10 movies that are similar to title of given moive
    moive_indices = [i[0] for i in sig_scores[1:11]]

    return movie_df_merge['original_title'].iloc[moive_indices]
    
```

```python
#testing our content based recommendation system with the film Spy Kids
get_recommendation("Spy Kids")
```
